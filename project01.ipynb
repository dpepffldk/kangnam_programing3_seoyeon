## 2. 데이터 탐색 (EDA)

#### 분석전 데이터 탐색을 통해 데이터를 어떤 방향으로 분석하고 도출할지 찾아낸다.
- 각 columns의 데이터 타입 알아보기
- describe()를 통해 데이터 값을 통계내기
- 시각화를 통해 데이터를 이해하기

```python
# 먼저 1번 데이터에 대한 데이터 탐색을 실행했다.
# 1. 유형별, 지역별 주택 매매 수급 동향

h_supply_demand.info()
```

데이터의 전반적인 정보를 보았을 때 columns는 총 10개, rows는 총 72개로 이루어져있고

memory 사용량은 5.8 KB이다.

또 주택유형별, 지역별 column에는 object 타입이 존재하고,

나머지 연도(2013-2020)를 가르키는 column에는 float 타입의 데이터가 존재한다.

```python
h_supply_demand.describe().T
```

2013년 부터 2020년까지 각 연도별 수급 매매 현황의 통계이다.

100%를 기준으로 나타내기 때문에 min과 max의 차이가 크지 않다.

```python
h_supply_demand.select_dtypes(object).columns
```

```python
# 데이터를 이용해 시각화 하기전 column을 뽑아낼 수 있도록 이름 바꾸기
h_supply_demand = h_supply_demand.rename(columns={h_supply_demand.columns[0]:'housing_type',
                                                   h_supply_demand.columns[1]:'area',
                                                   h_supply_demand.columns[2]:'y2013',
                                                   h_supply_demand.columns[3]:'y2014',
                                                   h_supply_demand.columns[4]:'y2015',
                                                   h_supply_demand.columns[5]:'y2016',
                                                   h_supply_demand.columns[6]:'y2017',
                                                   h_supply_demand.columns[7]:'y2018',
                                                   h_supply_demand.columns[8]:'y2019',
                                                   h_supply_demand.columns[9]:'y2020'})
h_supply_demand
```

```python
h_supply_demand.area.nunique()
```

```python
h_supply_demand.area.value_counts()
```

area의 각 유니크의 value_counts 값이 같은 것으로 보아서,

여기서 object타입 데이터인 area와 housing_type은 단지 구분을 위한 인덱스 역할임을 알 수 있다.

```python
# 중요한 값에 대한 데이터는 이부분이다.
h_supply_demand.iloc[:, 2:10]
```

```python
# 전체적인 데이터를 보기위해 지역별이 아닌 주택 형태별 매매 수급 평균보기
h_housing_mean = (h_supply_demand.groupby('housing_type').agg('mean'))
h_housing_mean
```

```python
# 기본 막대 그래프 그리기
h_housing_mean.iloc[3].plot(kind='bar')
```

```python
# 그다음은 2번 데이터에 대한 데이터 탐색을 실행했다.
# 2. 유형별, 지역별 주택 매매 가격 지수

h_price.info()
```

데이터의 전반적인 정보를 보았을 때 columns는 총 20개, rows는 총 72개로 이루어져 있으며,

memory 사용량은 11.4 KB이다.

주택유형별, 지역별 column에는 object 타입이 존재해야 하고

연도(2003-2020)를 가르키는 column에는 float 타입의 데이터가 존재해야 하는데

2003년부터 2011년 값에는 수치형 데이터가 들어가 있는데도 object타입으로 표기되어 있다.

데이터를 살펴본 결과, 수치표기가 없는 즉 null값 자리에 '-'가 들어가서 object타입으로 인식을 했다.

그러므로 결측치가 있는 해당 년도는 해당 컬럼의 mean이나 count에 크게 영향을 줄 수 있으므로

2003년부터 2011년도는 제외하고 2012년도부터 2020년도 데이터만 사용한다.

```python
# 시각화를 하기전 column을 뽑아낼 수 있도록 이름 바꾸기

h_price = h_price.rename(columns={h_price.columns[0]:'housing_type',
                                h_price.columns[1]:'area',
                                h_price.columns[2]:'y2003',
                                h_price.columns[3]:'y2004',
                                h_price.columns[4]:'y2005',
                                h_price.columns[5]:'y2006',
                                h_price.columns[6]:'y2007',
                                h_price.columns[7]:'y2008',
                                h_price.columns[8]:'y2009',
                                h_price.columns[9]:'y2010',
                                h_price.columns[10]:'y2011',
                                h_price.columns[11]:'y2012',
                                h_price.columns[12]:'y2013',
                                h_price.columns[13]:'y2014',
                                h_price.columns[14]:'y2015',
                                h_price.columns[15]:'y2016',
                                h_price.columns[16]:'y2017',
                                h_price.columns[17]:'y2018',
                                h_price.columns[18]:'y2019',
                                h_price.columns[19]:'y2020'})

h_price
```

```python
h_price.describe().T
```

2012년 부터 2020년까지 각 연도별 주택 매매 가격 지수의 통계이다.

2017년을 기준(100%)으로 나타내기 때문에 min과 max의 차이가 크지 않다.

```python
# 전체적인 데이터를 보기위해 지역별 매매 가격 지수의 평균보기
hp_area_mean = (h_price.groupby('area').agg('mean'))
hp_area_mean
```

```python
# 기본 막대 그래프 그리기 (전국, 평균, 2017년이 기준점)
hp_area_mean.iloc[12].plot(kind='bar')
```

```python
# 그다음은 3번 데이터에 대한 데이터 탐색을 실행했다.
# 3. 지역별 출산율

birth.info()
```

memory 사용량은 2.9 KB이다.

```python
# 기본 막대 그래프 그리기 (전국, 연도별, 출산수)
# 전체적인 데이터를 보기위해 지역별 매매 가격 지수의 평균보기
birth.iloc[0:1].plot(kind='bar')
```
