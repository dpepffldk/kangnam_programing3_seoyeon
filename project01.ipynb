## 3. 데이터 분석

#### 데이터 분석을 통해 데이터를 용이하게 변환하고 데이터로부터 의미 있는 결과를 도출한다.
- 데이터 전처리 및 정제
- 시각화를 통해 데이터를 이해하기

```python
# 데이터 전처리 및 정제
# 지역별 주택 매매 가격 지수에서는 전국 주택 매매지수 2003년도 ~ 2020년도 데이터만 가져온다.

h_price_entire_region = h_price.iloc[0, 1:20]

data1 = pd.DataFrame(h_price_entire_region)
data1 = data1.transpose()
data1
```

```python
# 데이터 병합을 위해 column 이름 바꾸기

data1 = data1.rename(columns={'area':'items'})
data1
```

```python
# 인구동향 조사에서는 출산율 2000년도 ~ 2020년도 데이터만 가져온다.

popul_birth = popul.iloc[6, 0:22]

data2 = pd.DataFrame(popul_birth)
data2 = data2.transpose()
data2
```

```python
# 사용할 두 개의 데이터 합치기

result = pd.concat([data1, data2])
result
```

인덱스를 0과 6의 이름을 바꿔서 사용할 것이라 items column은 사용하지 않음.

또 출산율은 2000년도 ~ 2002년도까지 데이터가 존재하지만

주택매매지수는 2003년도부터 데이터가 존재하므로 두 개의 데이터를 합쳤을 때 na값이 생긴다.

그래서 이 컬럼들은 빼고 다시 정의한다.

```python
result = result.iloc[:, 1:19]
result
```

```python
# 불필요한 데이터 컬럼을 제외했기에 다시 transpose()사용하여 row와 col의 위치를 바꿔준다.

result = result.transpose()
result
```

```python
# 컬럼 이름 바꾸기

result = result.rename(columns={result.columns[0]: 'price',
                                result.columns[1]: 'birth'})
result
```

여기서 인덱스는 연도 자체이다.

price는 전국의 주택매매 가격지수에 대한 데이터이고,

birth는 전국의 출산율에 대한 데이터 이다.

데이터 전처리 및 정제의 과정은 끝났다.

이제 이 데이터를 가지고 시각화를 해서 분석하고 이해하려고 한다.

```python
# 시각화 1 - 막대그래프 matplotlib 이용
# 연도별 주택 매매 가격 지수 변동
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(10, 8))
plt.barh(result.price.index, result.price)
plt.title('Annual Home Sales Price Index', fontsize=20)
plt.xlabel('price', fontsize=18)
plt.ylabel('year', fontsize=18)
```

연도별 전국의 주택 매매 가격 지수를 살펴본 결과

주택 매매 가격 지수는 2003년과 2011년을 제외하면 해마다 올라가고 있는 것을 알 수 있다.

2003년과 2011년은 전년도에 비해 더 큰폭의 상승을 보여주고 있다.

```python
# 시각화 2 - 막대그래프 seaborn 이용
# 연도별 출산율 비교
import seaborn as sns

fig, ax = plt.subplots(figsize=(10, 8))
sns.barplot(x=result.birth.index, y=result.birth)
plt.title('Annual birth rate', fontsize=20)
plt.xlabel('year', fontsize=18)
plt.ylabel('birthrate', fontsize=18)
```

연도별 출산율을 살펴본 결과

2003년에서 2015년까지 증가 감소를 반복하다가

2015년부터 2020년까지 급격한 감소를 보이고 있다.

```python
# 두 변수관의 산포도
fig, ax = plt.subplots(figsize=(8, 8))
result.plot.scatter(
    x='price',
    y='birth',
    alpha=.5,
    ax=ax
)
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(16, 4))
result.plot(ax=ax)
plt.title('Annual birth rate and Home Sales Price Index', fontsize=20)
```

산포도와 선그래프 모두 두 변수의 값의 범위가 차이가 나서 제대로된 결과를 볼 수 없다.

변수 선택을 할 때 이점도 고려한다.

데이터를 수집할 때 이부분을 고려하지 못하여서 제대로 된 도출점을 낼 수 가 없다.

그래서 우선은 선그래프 비교의 예시를 보여주기 위해

수집한 데이터 중 2번 인구동향 조사(출생,혼인,사망,조혼 등) 데이터의 출생아수, 혼인건수, 이혼건수에 대해 비교해보려고 한다.

```python
# 앞서 데이터 탐색 부분에서 결측치를 제거한 2번 데이터를 가져와 살핀다.
popul
```

```python
# 출생아수 뽑아내기
Births = popul.iloc[0, 0:22]

data3 = pd.DataFrame(Births)
data3 = data3.transpose()
data3
```

```python
# 혼인건수 뽑아내기
marry = popul.iloc[8, 0:22]

data4 = pd.DataFrame(marry)
data4 = data4.transpose()
data4
```

```python
# 이혼건수 뽑아내기
divorce = popul.iloc[10, 0:22]

data5 = pd.DataFrame(divorce)
data5 = data5.transpose()
data5
```

```python
# 3개의 데이터 합치기
result2 = pd.concat([data3, data4, data5])
result2
```

```python
# items 제거
result2 = result2.iloc[:, 1:22]
result2
```

```python
result2 = result2.T
result2
```

```python
# 여기서 사용할 데이터 출생아수, 혼인건수, 이혼건수에 대해 선그래프를 그리기 위해
# 해당 col 이름을 바꾼다.

result2 = result2.rename(columns={result2.columns[0]: 'Birth',
                              result2.columns[1]: 'Marry',
                              result2.columns[2]: 'Divorce'})
result2
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(16, 4))
result2.plot(ax=ax)
plt.title('Birth vs Marriage vs Divorce', fontsize=20)
```

아까보다는 유의미한 결과를 볼 수 있는 선 그래프가 만들어졌다.

해가 지날수록 출생아 수와 혼인수는 감소되는 동향을 보인다.

그러나 이혼수는 크게 차이가 없는 것으로 보인다.

또 혼인수가 많아지면 출생수도 많아지는 경향을 보인다.

그리고 출생수가 높았을 때 이혼수가 출생수가 낮았을 때 이혼수보다 낮은 것을 알 수 있다.

```python
# 2번째 데이터 기상청에서 2020년 1월부터 12월까지
# 서울의 기온, 풍속, 습도, 증기압, 이슬점온도, 현지기압, 해면기압, 지면온도 등을 1시간마다 측정한 데이터를 이용해
# 시각화를 해보자

weather_info = pd.read_csv('data/seoul_weather.csv', encoding='cp949')
weather_info.dtypes
```

```python
weather_info.head(10)
```

```python
# 지점은 위의 1번째 미세먼지 데이터와 마찬가지로 서울 측정소를 가르키는 고유번호이다.
# 컬럼을 부르기 쉽게 데이터 컬럼의 이름을 변경하기

weather = weather_info.rename(columns={weather_info.columns[0]: 'Location',
                                       weather_info.columns[1]: 'Reported_date', #날짜
                                       weather_info.columns[2]: 'Temperature', #기온
                                       weather_info.columns[3]: 'Windspeed', #풍속
                                       weather_info.columns[5]: 'Humidity', #습도
                                       weather_info.columns[8]: 'local_pressure', #현지기압
                                       weather_info.columns[9]: 'sea_pressure' #해면기압
                                      })

weather
```

```python
# 서울의 미세먼지 측정소 108번을 이해하기 쉽게 seoul로 변경
# 특정 컬럼의 해당 값을 변경

weather.loc[(weather.Location == 108), 'Location'] = '서울'
weather
```

```python
# 2020년 8월 서울의 기온과 습도 그래프
weather['Reported_date'] = pd.to_datetime(weather['Reported_date'])
spot = '서울'

data02 = weather[(weather['Location'] == spot) & (weather['Reported_date'].dt.month == 8)]
ax = data02.plot('Reported_date', 'Temperature', figsize = (12,6))
plt.title('8월 서울 기온, 습도 추이', fontsize = 16)
data02.plot('Reported_date', 'Humidity', secondary_y=True, ax=ax)
plt.show()
```

8월의 서울 기온과 습도를 비교해보면 반비례가 보이는 것을 알 수 있다.

습도가 높아지면 기온이 낮음을 보이고, 습도가 낮아지면 기온이 높음을 알 수 있다.

```python
# 상관계수를 그릴 컬럼 선택 : 기온, 풍속, 습도, 현지기압, 해면기압

heat_map = weather[
    ['Temperature', 'Windspeed', 'Humidity', 'local_pressure', 'sea_pressure']
].dropna().copy()
```

```python
# 히트맵, 상관계수, seaborn 이용
import seaborn as sns

plt.figure(figsize = (10,10))
plt.title('Correlation', fontsize=16)
sns.heatmap(data = heat_map.corr(),
            annot = True,
            fmt = '.2f', linewidths = .5, cmap = 'coolwarm',
           vmin = -1, vmax = 1, center = 0)
plt.show()
```

해당 그래프를 확인해 보면 상관계수는 1에 가까울수록 상관이 있음을 보인다.

여기서는 현지기압과 해면기압이 관계가 있다고 나오고

기온과 풍속이 아주 조금 관계가 있고, 기온과 습도가 어느정도의 관계를 보이는 것을 알 수 있다.
